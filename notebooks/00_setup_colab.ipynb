# Colab Setup

Use this notebook at the start of a Colab session to mount Google Drive, move into the repository, install dependencies, and create the expected directory structure.

## 1. Mount Drive and Enter Project

```python
from google.colab import drive
from pathlib import Path
import os
import sys

drive.mount('/content/drive', force_remount=False)

project_path = Path('/content/drive/MyDrive/xai-dark-matter-localization')
os.chdir(project_path)
if str(project_path) not in sys.path:
    sys.path.insert(0, str(project_path))

print(f'Working directory: {Path.cwd()}')
```

## 2. Install Dependencies

```python
import subprocess
import sys

subprocess.check_call([sys.executable, '-m', 'pip', 'install', '-r', 'requirements.txt'])
```

## 3. Create Directories

```python
import subprocess
import sys

subprocess.check_call([sys.executable, '-m', 'src.setup_project'])
```

## 4. Check Environment

```python
from pathlib import Path
import os

print('DATA_ROOT:', os.getenv('DATA_ROOT', './data'))
print('TNG_API_KEY configured:', bool(os.getenv('TNG_API_KEY')))
for path in [
    Path('data/raw/tng/metadata_raw'),
    Path('data/processed/TNG-DM-XAI-v1'),
    Path('outputs'),
    Path('results/xai'),
    Path('results/analysis'),
]:
    print(path, 'exists=', path.exists())
```
